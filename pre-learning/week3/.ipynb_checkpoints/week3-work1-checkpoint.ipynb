データが読み込めたら、説明変数を下のようなPandasのDataFrame型変数Xに格納してください。列の名前も以下を参考にしてください。


X =

```python
from sklearn.datasets import load_iris
import pandas as pd
from pandas import DataFrame
import numpy as np

#irisデータセットの読み込み
iris = load_iris()

# 説明変数をPandasのDataFrame型変数Xに格納
X = DataFrame(iris.data, columns = iris.feature_names)

# 目的変数をPandasのDataFrame型変数yに格納し、カラム名を"Species"に変更
y = DataFrame(iris.target, columns = ["Species"])

X
```

## 【課題2】データの結合
そして、これらX,yを結合させたDataFrame型変数dfを作成してください。これ以降はこのdfを使っていきます。

```python
#X,yを結合させたDataFrame型変数dfを作成
df = pd.concat([X, y], sort=False, axis=1)
df
```

## 【課題6】図の作成
データは可視化して確認することが大切です。図を作成していきましょう。Matplotlibやseabornを使ってください。

- ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）
- 特徴量を一つ選び、boxplotを使ってラベルごとのデータの分布を可視化する
- 特徴量を一つ選び、violinplotを使ってラベルごとのデータの分布を可視化する


boxplotとviolinplotは似た図を表示します。これらはどう違い、それぞれどのような利点があるのか説明してください。

### 1.ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）

```python
import matplotlib.pyplot as plt
import seaborn as sns

#ラベルづけ
label = ['setosa', 'versicolor', 'virginica']
# 円グラフを表示
plt.pie(df["Species"].value_counts(), autopct="%1.1f%%", labels = label)

plt.show()
```

### 2.特徴量を一つ選び、boxplotを使ってラベルごとのデータの分布を可視化する

```python
sns.set()

#データの作成
setosa_sepallength = df[df["Species"] ==  0]["sepal length (cm)"]
versicolor_sepallength = df[df["Species"] ==  1]["sepal length (cm)"]
virginica_sepallength = df[df["Species"] ==  2]["sepal length (cm)"]

# 新規のウィンドウを描画
fig = plt.figure()

# サブプロットを追加
ax = fig.add_subplot(1, 1, 1)

#データをセット
ax.boxplot([setosa_sepallength, versicolor_sepallength, virginica_sepallength], labels=label)

#x軸ラベル名
ax.set_xlabel('Species')

#y軸ラベル名
ax.set_ylabel('sepal length (cm)')

# 縦軸の範囲を指定
ax.set_ylim(3, 9)

plt.show()
```

### 3.特徴量を一つ選び、violinplotを使ってラベルごとのデータの分布を可視化する

```python
sns.set()

#データの作成(がく片の長さ)
setosa_sepallength = df[df["Species"] ==  0]["sepal length (cm)"]
versicolor_sepallength = df[df["Species"] ==  1]["sepal length (cm)"]
virginica_sepallength= df[df["Species"] ==  2]["sepal length (cm)"]

# 新規のウィンドウを描画
fig = plt.figure()

# サブプロットを追加
ax = fig.add_subplot(1, 1, 1)

#データをセット
ax.violinplot([setosa_sepallength, versicolor_sepallength, virginica_sepallength])

# x軸への目盛り設定位置の指定
ax.set_xticks([1, 2, 3])

# データラベル名
ax.set_xticklabels(label)

#x軸ラベル名
ax.set_xlabel('Species')

#y軸ラベル名
ax.set_ylabel('sepal length (cm)')

# 縦軸の範囲を指定
ax.set_ylim(3, 9)

plt.show()
```

 ### ※boxplotとviolinplotは似た図を表示します。これらはどう違い、それぞれどのような利点があるのか説明してください。

- boxplot  
2種類のカテゴリデータを含んだ数値データのばらつきを可視化するのに適しています。  
特に、四分位点と中央値にフォーカスしたい場合に利点があります。

- violinplot  
boxplotと同様に，２種類のカテゴリデータを含んだ数値データの違いを可視化し、各数値データを分布として表現します。  
特に、分布の形や多峰性を見たい場合に利点があります。


## 【課題7】特徴量間の関係の確認
特徴量間の関係を見ていきます。

- 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
- 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
- ４つの特徴量に関する相関係数行列を作る
- 相関係数行列のヒートマップを作る


### 1.縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る

```python
#データの作成（がく片の幅）
setosa_sepalwidth = df[df["Species"] ==  0]["sepal width (cm)"]
versicolor_sepalwidth = df[df["Species"] ==  1]["sepal width (cm)"]
virginica_sepalwidth= df[df["Species"] ==  2]["sepal width (cm)"]

#setosaの[がく片の長さ*がく片の幅]散布図を出力（色は赤）
plt.scatter(setosa_sepallength, setosa_sepalwidth, c="r", marker="o",)
#virginicaの[がく片の長さ*がく片の幅]散布図を出力（色は青）
plt.scatter(versicolor_sepallength, versicolor_sepalwidth, c="b", marker="o",)
#virginicaの[がく片の長さ*がく片の幅]散布図を出力（色は緑）
plt.scatter(virginica_sepallength, virginica_sepalwidth, c="g", marker="o",)
```



### 2.全ての組み合わせの散布図を一度に表示させる、散布図行列を作る

```python
#seabornを用いたirisデータの読み込み
df2 = sns.load_dataset("iris")

#散布図行列をセットする
sns.pairplot(df2, hue="species", height=3)
plt.show()
```

### 3.４つの特徴量に関する相関係数行列を作る

```python
'''
<memo>
相関係数の算出方法の指定: 引数method
corr()メソッドでは引数methodで相関係数の算出方法を指定できる。

以下の3種類から選択する。

'pearson': ピアソンの積率相関係数（デフォルト）
'kendall': ケンドールの順位相関係数
'spearman': スピアマンの順位相関係数
'''

#相関係数行列を作成
corr_iris = df2.corr(method="pearson")
corr_iris
```

### 4.相関係数行列のヒートマップを作る

```python
#ヒートマップを作成
sns.heatmap(corr_iris,
            vmin=-1.0,
            vmax=1.0,
            center=0,
            annot=True, # True:格子の中に値を表示
            fmt='.1f',
            xticklabels=corr_iris.columns.values,
            yticklabels=corr_iris.columns.values
           )
plt.show()
```
