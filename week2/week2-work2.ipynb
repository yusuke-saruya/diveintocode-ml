## 【課題1】1次元配列を作ろう
0~1の範囲で0.01ごとにデータをNumPy配列で作成してください。

データは[0, 0.01, 0.02・・・0.98, 0.99, 1]という形式の配列になります。

```python
import numpy as np

#0〜1までの公差0.01の等差数列の生成
data = np.arange(0, 1.01, 0.01)

#出力
print(data)
```

## 【課題2】2次元配列を作ろう
以下のような単純な 
$
    y=\frac{1}{2}x+1
$
 の関数について考えます。
 
 xの範囲を[0,1]、xの刻み幅を0.01としてyの値を計算しなさい。そして計算結果をNumPy配列の0列目にx、1列目にyの値がくるように2次元のarrayで表現しなさい。
計算する際は、上の【課題1】で作成したNumPy配列を用いること。

```python
"""
y=1/2x + 1 関数

Parameter
------------
Input:
    x:int
    
Return:
    一次関数 y = x / 2 + 1
    
"""
def f(x):
    return x / 2 + 1

#【課題1】で作成した配列を2次元にreshapeし、転置する
data_transpose = np.reshape(data, (1, len(data))).T

#0列目にx、1列目にyの値をセット
data_xy = np.insert(data_transpose, 1, f(data), axis=1)
print(data_xy)
```

## 【課題3】勾配を求めよう
機械学習では関数の勾配がとても重要です。関数の勾配を求める考えを知っておくことはこの4ヵ月の間で一番重要かもしれません。

関数の勾配とは、中学数学でいう所の「傾き」です。これ以降は言葉を勾配で統一します。勾配を求める公式は以下で表現されることを思い出しましょう。
$$
    \frac{yの変化量}{xの変化量}
$$
上の公式を使って、各x座標における勾配を求めてください。ここでも上の【課題1】で作成したNumPy配列を用いて計算してください。(1次元のlist等に変換しないこと)

注意

- 厳密に各点における勾配を求めるには極限（微分）を利用する必要がありますが、ここでは各x座標=隣り合うx座標の間隔だと思ってください。
- 勾配はx座標の間隔を使うので、xの点の数に対して一つ少なくなります。
- 普通に求めると100個の勾配が出てきますが、残りの一つの勾配は、自分で考えて値を埋めてください。

```python
"""
勾配の関数

Parameter
------------
Input:
    x:int
    y:int
    
return:
    (yの変化量)/(xの変化量)
"""
def gradient(x, y):
    return y / x


#各x座標における勾配
for i in range(100):
    #x,yの変化量
    change_x = data_xy[i+1][0] - data_xy[i][0] 
    change_y = data_xy[i+1][1] - data_xy[i][1] 
    print("x=" + str(data_xy[i][0]) + "　：　" + str(gradient(change_x, change_y)))

#残りの一つの勾配は自分で考える
selfdata_x = 1.02
selfdata_y = f(selfdata_x)
#変化量
selfchange_x = selfdata_x - data_xy[100][0] 
selfchange_y = selfdata_y - data_xy[100][1] 
#101番目の勾配
print("x=1　：　" +  str(gradient(selfchange_x, selfchange_y)))
```

## 【課題4】配列、勾配、グラフを作ろう
次に$y=x^2$ について考えます。（xの範囲は[0,1]）

【課題2】、【課題3】と同様に、2次元配列を作り、勾配を求めてください。

さらに、求めた勾配をMatplotlibを用いてグラフで描画してください。

```python
import matplotlib.pyplot as plt
%matplotlib inline

"""
y=x^2 関数

Parameter
------------
Input:
    x:int
    
Return:
    二次関数 y = x^2
    
"""
def f2(x):
    return x ** 2

#0列目にx、1列目にyの値をセット
data_xy2 = np.insert(data_transpose, 1, f2(data), axis=1)

"""
勾配の関数

Parameter
------------
Input:
    x:int
    y:int
    
return:
    (yの変化量)/(xの変化量)
"""
def gradient2(x, y):
    return y / x

#各x座標における勾配
list_gradient2 = []
for i in range(100):
    #x,yの変化量
    change_x2 = data_xy2[i+1][0] - data_xy2[i][0] 
    change_y2 = data_xy2[i+1][1] - data_xy2[i][1] 
    print("x=" + str(data_xy2[i][0]) + "　：　" + str(gradient2(change_x2, change_y2)))
    #リストに加える
    list_gradient2.append(gradient2(change_x2, change_y2))

#残りの一つの勾配は自分で考える
selfdata_x2 = 1.02
selfdata_y2 = f2(selfdata_x2)
#変化量
selfchange_x2 = selfdata_x2 - data_xy2[100][0] 
selfchange_y2 = selfdata_y2 - data_xy2[100][1] 
#101番目の勾配
print("x=1　：　" +  str(gradient2(selfchange_x2, selfchange_y2)))
list_gradient2.append(gradient2(selfchange_x2, selfchange_y2))

plt.plot(data, list_gradient2)
plt.show()
```

## 【課題5】最小値の求め方を説明しよう
上で求めた勾配を使用して、$y=x^2$の最小値を求めるにはどうしたらいいでしょうか？ 
例えば谷を降りる例を考えましょう。この場合、坂の傾きの方向に歩いていけば谷の最下点（最小値）にたどり着くことは直感的にわかると思います。要するに、谷の現在地から最下点にたどり着くには、歩くたびにその地点での勾配を求めてその方向に進んでいくということを繰り返して行けば良いわけです。このように谷の最小値を求めるために、地道に現在地の勾配を計算し、その方向に進んでいく方法を勾配降下法と言います。当然最下点に近くにつれ、坂の傾斜は緩やかになるので、勾配もだんだんと0に近づいて行きます。

今回は谷の例で考えましたが、谷の形が複雑な数式で表現されるような関数であっても同じことです。


上記を参考に、勾配を使用して$y=x^2$ の最小値の求める方法を説明してください。

勾配降下法というワードを使用して、記述してください。実装ではなく、アルゴリズムのフローチャートを記述してください。図でも文章でもどちらでも構いません。



## 回答（課題5）
$y=x^2$を書き直し、$f(x) = x^2$とする。  
ここで$f(x)$を微分すると、$\frac{d}{dx}f(x) = 2x$ となり、二次関数$f(x)$の増減表は下記の通りとなる。

|xの範囲  |$\frac{d}{dx}f{(x)}$の符号  |$$f{(x)}$$の増減  |
|---|---|---|
|$$x<0$$  |$$-$$  |右下がり  |
|$$x=0$$  |$$0$$  |  |
|$$x>0$$  |$$+$$  |右上がり  |

この増減表に従ってxの値を増減させることにより、最小値を求めることができる。  
例えば、下記グラフ( $f(x)=x^2$ )の通り,x=2の場合、xを減らして行くことにより、x=0の時に最小値となることがわかる。


```python
import matplotlib.pyplot as plt  
import numpy as np 

# -3から3までの公差0.01の等差数列の生成
x = np.arange(-3, 3, 0.01) 
y = x**2

plt.plot(x,y)
plt.scatter(2,4, c="r")
plt.scatter(0,0, c="r")
plt.plot([2,2], [4,0], c="r", linestyle="dashed")
plt.plot([0,2], [0,0], c="r", linestyle="dashed")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show() 
```

ここで、学習率を$\eta$とすると、xの値を更新する式は下記の通りとなる。 $$x := x - \eta\frac{d}{dx}f(x)$$
ここで例として、$x =2$、学習率$\eta=0.3$とすると、勾配降下法を用いて下記のように$x=0$へ向かって収束し、最小値を求めることができる。  
$x := 2 - 0.4(2\times2) = 0.4$  
$x := 0.4 - 0.4(2\times0.4) = 0.08$  
$x := 0.08 - 0.4(2\times0.08) = 0.016$

## 【課題6】最小値を求めよう
最後に具体例を計算してもらって締めくくりとしましょう！  
$y=x^2+5x$ の最小値を勾配降下法ではなく、微分法を使って求めてください。

## 回答（課題6）
$f(x)=x^2+5x$とすると、  
$\frac{d}{dx}f(x)=2x+5$  

$\frac{d}{dx}f(x)=0$となるときに、f(x)が最小値となるため、  
$2x+5=0$  
$2x=-5$  
$x=-\frac{5}{2}$  

よって$f(x)$の最小値は、  
$f(-\frac{5}{2})=(-\frac{5}{2})^2 + 5\cdot(-\frac{5}{2})$  
$\hspace{27pt}= \frac{25}{4} - \frac{25}{2}$  
$\hspace{27pt}= \frac{25-50}{4}$  
$\hspace{27pt}= -\frac{25}{4}$
